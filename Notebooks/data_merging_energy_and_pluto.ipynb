We will join the dataset using BBL. But since the format is still different, we have to take care of that.

# Fix BBL Format and Data Type

```python
# Remove dashes and keep as string
energy_df['NYC Borough, Block and Lot (BBL)'] = energy_df['NYC Borough, Block and Lot (BBL)'].str.replace("-", "")
energy_df.head(5)
```

```python
# Convert the data type of the 'bbl' column in 'pluto_df' to string
pluto_df['bbl'] = pluto_df['bbl'].astype(int)
pluto_df['bbl'] = pluto_df['bbl'].astype(str)
pluto_df.info()
```

```python
# Rename the column: make it lowercase and replace spaces and remove non-numeric characters
def clean_column_names(column):
    column = column.lower()  # Convert to lower case
    column = column.replace(' ', '_')  # Replace spaces with underscores
    column = ''.join([char for char in column if char.isalnum() or char == '_'])  # Keep alphanumeric and underscore
    return column

# Apply the function to the DataFrame columns
energy_df.columns = [clean_column_names(col) for col in energy_df.columns]
energy_df.info()
```

# Merge Data

```python
merged_data = pd.merge(energy_df, pluto_df, left_on="nyc_borough_block_and_lot_bbl", right_on="bbl", how='outer', indicator=True)
merged_data.shape
```

```python
inner_joined_data = merged_data[merged_data['_merge'] == 'both']
inner_joined_data.shape
```

```python
inner_joined_data
```

```python
inner_joined_data.to_csv('merged_data_energy_pluto.csv')
```

```python
# Check null values
# Calculate the percentage of null values in each column
null_percentage = (inner_joined_data.isnull().sum() / inner_joined_data.shape[0] * 100).to_frame(name='Null Percentage')

# Create a heatmap to visualize null percentages
plt.figure(figsize=(2, 15))
sns.heatmap(null_percentage, annot=True, cmap='Reds', cbar=True, fmt=".2f")
plt.title("Heatmap of Null Percentages in Each Column")
plt.xlabel("Columns")
plt.show()
```
